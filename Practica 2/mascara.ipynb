```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

comida1 = cv2.imread(r'./comida/Entrenamiento1.jpg')
comida1 = cv2.cvtColor(comida1, cv2.COLOR_BGR2RGB)


plt.imshow(comida1)
```

```python
#Mascara de los huevos
mask1 = np.zeros(comida1.shape[:2],dtype="uint8")

cv2.circle(mask1,(220,440),60,255,-1)
cv2.circle(mask1,(285,340),60,255,-1)
cv2.circle(mask1,(350,455),63,255,-1)
print(mask1.shape)
huevos = cv2.bitwise_and(comida1,comida1, mask=mask1)
plt.imshow(huevos)
```
